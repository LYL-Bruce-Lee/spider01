```python
import pymongo,sys
from string import punctuation
import charts

client=pymongo.MongoClient('localhost',27017)

ceshi=client['ceshi']
item_info=ceshi['item_infoY']

# db.getCollection('item_infoY').find({})


for i in item_info.find().limit(10):
    print(i['place'])


# 1、skip（2）:从数据集的第二条开始查询

# 2、limit(2) : 依次查出2条数据。

# 3、sort({age:1}) : 1.正序查询，-1倒叙查询。

# 4、count():结果集总数。


for i in item_info.find().limit(30):
    print(i['place'])

# for i in item_info.find():
#     try :
#         if not isinstance(i, list):
#             if i['place'] == '------':
#                 place=['不明']
#             else:
#                 place=i['place'].split('-')
# #             print(place)
#             item_info.update({'_id':i['_id']},{'$set':{'place':place}})
#     except Exception:
#         print(i)

# series=[{'name':'os x',
#         'data':[11],
#         'type':'column'},
#        {'name':'ubuntu',
#        'data':[8],
#        'type':'column'},
#         {'name':'tohers',
#         {'data':[29],
#         'type':'column'}}]
series=[{'name':'os x',
        'data':[11],
        'type':'column'},
       {'name':'ubuntu',
       'data':[8],
       'type':'column'},
        {'name':'tohers',
        'data':[29],
        'type':'column'}]
series2=[{'name':'john','data':[5],'type':'column'},{'name':'hohn','data':[50],'type':'column'}]

charts.plot(series2,show='inline',options=dict(title=dict(text="charts are awesome!!!")))

place_list=[]
try:
    for i in item_info.find():
        if len(i['place'])>1:
            place_list.append(i['place'][1])
except Exception as e:
    print(len(i))
    print('wrong!\n'+str(e)+'\n'+str(i))
    pass
place_index=list(set(place_list))
print(place_index)

post_times=[]
for index in place_index:
    post_times.append(place_list.count(index))
print(post_times)

def data_gen(types):
    length=0
    if length<=len(place_index):
        for place,times in zip(place_index,post_times):
            data={
                'name':place,
                'data':[times],
                'type':types
            }
            yield data
            length+=1

data_gen('column')

for i in data_gen('column'):
    print(i)

series2=[data for data in data_gen('column')]
print(series2)
charts.plot(series2,show='inline',options=dict(title=dict(text="3月28日58同城北京交易数据!!!")))

```
